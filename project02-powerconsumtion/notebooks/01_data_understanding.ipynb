## 7. Tipe Data & Konversi

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
print(f'Tipe data Datetime: {df["Datetime"].dtype}')
print(f'Range waktu: {df["Datetime"].min()} s.d. {df["Datetime"].max()}')
```

## 8. Target Prediksi

```python
target_cols = ['PowerConsumption_Zone1', 'PowerConsumption_Zone2', 'PowerConsumption_Zone3']
print('=== Target Prediksi ===')
for col in target_cols:
    print(f'{col}: mean={df[col].mean():.2f}, min={df[col].min():.2f}, max={df[col].max():.2f}')
```
