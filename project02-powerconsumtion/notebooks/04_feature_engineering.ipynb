```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.dpi'] = 110

INPUT_PATH = '../data/processed_data.csv'
df = pd.read_csv(INPUT_PATH)
df['Datetime'] = pd.to_datetime(df['Datetime'])
df = df.sort_values('Datetime').reset_index(drop=True)
print(f'Loaded: {df.shape}')
```

```python
df['Hour'] = df['Datetime'].dt.hour
df['Day'] = df['Datetime'].dt.day
df['Month'] = df['Datetime'].dt.month
df['Weekday'] = df['Datetime'].dt.weekday
df['Weekend'] = (df['Weekday'] >= 5).astype(int)
print('Feature waktu ditambahkan.')
```

```python
df['Hour_sin'] = np.sin(2 * np.pi * df['Hour'] / 24)
df['Hour_cos'] = np.cos(2 * np.pi * df['Hour'] / 24)
print('Cyclic encoding selesai.')
```

```python
target_cols = ['PowerConsumption_Zone1', 'PowerConsumption_Zone2', 'PowerConsumption_Zone3']
lag_cols = [1, 2, 3, 6, 12, 144]
for col in target_cols:
    for lag in lag_cols:
        df[f'{col}_lag{lag}'] = df[col].shift(lag)
print(f'Lag features ditambahkan: {len(lag_cols)} lags x {len(target_cols)} targets')
```

```python
for col in target_cols:
    df[f'{col}_roll_mean_6'] = df[col].rolling(6).mean()
    df[f'{col}_roll_mean_12'] = df[col].rolling(12).mean()
    df[f'{col}_roll_std_6'] = df[col].rolling(6).std()
    df[f'{col}_roll_std_12'] = df[col].rolling(12).std()
print('Rolling statistics selesai.')
```

```python
df['Temp_Hum_Interaction'] = df['Temperature'] * df['Humidity']
print('Interaction features selesai.')
```

```python
df = df.dropna().reset_index(drop=True)
df.to_csv('../data/featured_data.csv', index=False)
print(f'featured_data.csv berhasil disimpan: {df.shape}')
```

## Kesimpulan

Feature baru yang ditambahkan:
- Waktu: Hour, Day, Month, Weekday, Weekend
- Cyclic: Hour_sin, Hour_cos
- Lag: 1, 2, 3, 6, 12, 144 (per 10 menit)
- Rolling: Mean & Std (window 6 & 12)
- Interaction: Temperature x Humidity
