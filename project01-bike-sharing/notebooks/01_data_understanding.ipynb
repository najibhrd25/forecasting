## 5. Data Structure

Pada bagian ini dilakukan analisis terhadap struktur internal dataset, termasuk tipe data setiap kolom dan kategorisasi feature sesuai Data Dictionary.

### 5.1 Informasi Dataset (`info()`)

Menampilkan ringkasan lengkap struktur dataset.

```python
# Informasi umum dataset
df.info()
```

**Interpretasi:**

- Tidak ada nilai null pada seluruh kolom (semua non-null count sama dengan 17.379).
- Kolom `dteday` bertipe `object` (string), perlu dikonversi ke tipe `datetime` pada tahap preprocessing.
- Kolom `temp`, `atemp`, `hum`, dan `windspeed` bertipe `float64`.
- Seluruh kolom lainnya bertipe `int64`.

### 5.2 Tipe Data per Kolom

```python
# Tipe data setiap kolom
print('=' * 50)
print('TIPE DATA SETIAP KOLOM')
print('=' * 50)
dtype_df = pd.DataFrame({
    'Kolom': df.dtypes.index,
    'Tipe Data': df.dtypes.values
}).reset_index(drop=True)
print(dtype_df.to_string(index=False))
```

### 5.3 Kategori Feature

Mengklasifikasikan seluruh kolom sesuai dengan Data Dictionary.

```python
# Klasifikasi feature sesuai Data Dictionary
FEATURE_GROUPS = {
    'Identifier'       : ['instant'],
    'Time Feature'     : ['dteday', 'yr', 'mnth', 'hr', 'weekday', 'season'],
    'Calendar Feature' : ['holiday', 'workingday'],
    'Weather Feature'  : ['weathersit', 'temp', 'atemp', 'hum', 'windspeed'],
    'User Feature'     : ['casual', 'registered'],
    'Target'           : ['cnt'],
}

print('=' * 55)
print('KLASIFIKASI FEATURE SESUAI DATA DICTIONARY')
print('=' * 55)
for group, cols in FEATURE_GROUPS.items():
    print(f'\n[{group}]')
    for col in cols:
        print(f'  - {col} ({df[col].dtype})')
```

**Interpretasi:**

Dataset memiliki 6 kelompok feature:

| Kelompok | Jumlah Kolom | Digunakan pada Model |
|----------|-------------|----------------------|
| Identifier | 1 | ❌ Tidak |
| Time Feature | 6 | ✅ Ya |
| Calendar Feature | 2 | ✅ Ya |
| Weather Feature | 5 | ✅ Ya |
| User Feature | 2 | ❌ Tidak (data leakage) |
| Target | 1 | 🎯 Target Prediksi |

Feature `instant`, `casual`, dan `registered` tidak akan digunakan sebagai input model. `casual` dan `registered` dikecualikan karena keduanya merupakan penyusun langsung dari `cnt` sehingga penggunaannya akan menyebabkan **data leakage**.

### 5.4 Konfirmasi Target Prediksi

```python
# Konfirmasi kolom target prediksi
TARGET_COL = 'cnt'

print(f'Target Prediksi : {TARGET_COL}')
print(f'Tipe Data       : {df[TARGET_COL].dtype}')
print(f'Nilai Min       : {df[TARGET_COL].min()}')
print(f'Nilai Max       : {df[TARGET_COL].max()}')
print(f'Nilai Rata-rata : {df[TARGET_COL].mean():.2f}')

# Verifikasi hubungan cnt = casual + registered
verifikasi = (df['cnt'] == df['casual'] + df['registered']).all()
print(f'\nVerifikasi cnt = casual + registered : {verifikasi}')
```

**Interpretasi:**

- Kolom `cnt` merupakan target prediksi dengan nilai integer non-negatif.
- Persamaan `cnt = casual + registered` terverifikasi untuk seluruh baris, mengonfirmasi bahwa `casual` dan `registered` **tidak boleh digunakan** sebagai feature input model.

## 6. Missing Value Analysis

Memeriksa keberadaan nilai yang hilang (missing value) pada seluruh kolom dataset. Identifikasi missing value sangat penting untuk menentukan strategi data cleaning yang tepat.

```python
# Analisis missing value
missing_count = df.isnull().sum()
missing_pct   = (df.isnull().sum() / len(df) * 100).round(4)

missing_df = pd.DataFrame({
    'Kolom'               : missing_count.index,
    'Jumlah Missing'      : missing_count.values,
    'Persentase Missing (%)': missing_pct.values
}).reset_index(drop=True)

print('=' * 55)
print('ANALISIS MISSING VALUE')
print('=' * 55)
print(missing_df.to_string(index=False))
print(f'\nTotal kolom dengan missing value : {(missing_count > 0).sum()}')
```

**Interpretasi:**

Dataset **tidak memiliki missing value** pada seluruh 17 kolom. Kondisi ini sesuai dengan informasi pada Data Dictionary dan menunjukkan kualitas data yang baik sehingga tidak diperlukan strategi imputasi. Tahap Data Cleaning selanjutnya dapat difokuskan pada aspek lain seperti tipe data dan penanganan outlier jika diperlukan.

## 7. Duplicate Analysis

Memeriksa keberadaan baris duplikat pada dataset. Baris duplikat dapat mengakibatkan bias pada model jika tidak ditangani.

```python
# Analisis duplikasi data
total_duplikat = df.duplicated().sum()
total_baris    = len(df)

print('=' * 45)
print('ANALISIS DUPLICATE DATA')
print('=' * 45)
print(f'Total baris               : {total_baris:,}')
print(f'Jumlah duplikat (seluruh kolom) : {total_duplikat}')
print(f'Persentase duplikat       : {total_duplikat / total_baris * 100:.4f}%')
```

```python
# Memeriksa duplikat berdasarkan kolom kunci time series (dteday + hr)
duplikat_kunci = df.duplicated(subset=['dteday', 'hr']).sum()

print('=' * 45)
print('DUPLIKAT BERDASARKAN KUNCI TIME SERIES')
print('Kunci: dteday + hr (tanggal + jam)')
print('=' * 45)
print(f'Jumlah duplikat kunci : {duplikat_kunci}')
```

**Interpretasi:**

Dataset **tidak mengandung baris duplikat**, baik secara keseluruhan maupun berdasarkan kombinasi kunci time series `dteday + hr`. Hal ini menunjukkan bahwa setiap jam memiliki satu observasi unik dan data sudah tertata dengan rapi. **Tidak diperlukan penghapusan duplikat.**

## 8. Statistical Summary

Menampilkan statistik deskriptif untuk memahami distribusi nilai pada setiap kolom numerik.

### 8.1 Statistik Deskriptif Umum

```python
# Statistik deskriptif seluruh kolom
df.describe().T
```

### 8.2 Statistik Tambahan (Median)

Menambahkan median yang tidak tersedia pada `describe()` bawaan.

```python
# Kolom numerik yang relevan untuk dianalisis
kolom_numerik = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

# Ringkasan statistik lengkap
stats = df[kolom_numerik].agg(['mean', 'median', 'min', 'max', 'std']).T
stats.columns = ['Mean', 'Median', 'Min', 'Max', 'Std Dev']
stats = stats.round(4)

print('=' * 65)
print('RINGKASAN STATISTIK KOLOM NUMERIK')
print('=' * 65)
print(stats.to_string())
```

**Interpretasi:**

**Kolom Target (`cnt`):**
- Rata-rata penyewaan per jam adalah sekitar 189 sepeda.
- Nilai minimum 1 dan maksimum 977 menunjukkan variasi yang sangat tinggi.
- Standar deviasi yang besar (~181) mengindikasikan pola yang sangat dipengaruhi waktu dan kondisi.

**Kolom Cuaca:**
- `temp` dan `atemp` memiliki nilai yang berdekatan, menunjukkan korelasi tinggi antar keduanya.
- `hum` (kelembapan) rata-rata sekitar 0.627 dalam skala 0–1.
- `windspeed` relatif rendah dengan rata-rata sekitar 0.19.

**Kolom User:**
- `registered` memiliki nilai rata-rata jauh lebih tinggi dari `casual`, menunjukkan pengguna terdaftar mendominasi total penyewaan.

### 8.3 Statistik Kolom Kategorikal

```python
# Statistik kolom kategorikal (nilai unik dan frekuensi)
kolom_kategorikal = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']

print('=' * 55)
print('RINGKASAN KOLOM KATEGORIKAL')
print('=' * 55)
for col in kolom_kategorikal:
    nilai_unik = sorted(df[col].unique())
    print(f'\n{col} | Nilai unik: {nilai_unik}')
    print(df[col].value_counts().sort_index().to_string())
```

**Interpretasi:**

- **`season`**: Data mencakup 4 musim dengan distribusi yang relatif merata.
- **`yr`**: Dua tahun data (0 = 2011, 1 = 2012).
- **`mnth`**: Data mencakup seluruh 12 bulan dalam setahun.
- **`hr`**: Data mencakup seluruh 24 jam (0–23).
- **`holiday`**: Mayoritas data adalah hari non-libur.
- **`weekday`**: Data terdistribusi merata di seluruh hari dalam seminggu.
- **`workingday`**: Sekitar 68% data merupakan hari kerja.
- **`weathersit`**: Mayoritas data berada pada kondisi cuaca cerah (nilai 1), dan tidak ada atau sedikit sekali data dengan cuaca ekstrem (nilai 4).

## 9. Validasi Data Dictionary

Membandingkan dataset dengan dokumentasi Data Dictionary untuk memastikan kesesuaian struktur, nama feature, tipe data, dan target prediksi.

### 9.1 Validasi Jumlah Feature

```python
# Jumlah feature sesuai dokumentasi
EXPECTED_ROWS = 17379
EXPECTED_COLS = 17

print('=' * 50)
print('VALIDASI DIMENSI DATASET')
print('=' * 50)
print(f'Jumlah baris  — Dokumentasi : {EXPECTED_ROWS:,} | Aktual : {len(df):,} | '
      f'Status: {"✅ Sesuai" if len(df) == EXPECTED_ROWS else "❌ Tidak Sesuai"}')
print(f'Jumlah kolom  — Dokumentasi : {EXPECTED_COLS}   | Aktual : {len(df.columns)}   | '
      f'Status: {"✅ Sesuai" if len(df.columns) == EXPECTED_COLS else "❌ Tidak Sesuai"}')
```

### 9.2 Validasi Nama Kolom

```python
# Nama kolom sesuai dokumentasi
EXPECTED_COLUMNS = [
    'instant', 'dteday', 'season', 'yr', 'mnth', 'hr',
    'holiday', 'weekday', 'workingday', 'weathersit',
    'temp', 'atemp', 'hum', 'windspeed',
    'casual', 'registered', 'cnt'
]

print('=' * 55)
print('VALIDASI NAMA KOLOM')
print('=' * 55)

kolom_aktual = list(df.columns)
for col in EXPECTED_COLUMNS:
    status = '✅ Ada' if col in kolom_aktual else '❌ Tidak Ada'
    print(f'  {status} : {col}')

kolom_tidak_terdokumentasi = [c for c in kolom_aktual if c not in EXPECTED_COLUMNS]
if kolom_tidak_terdokumentasi:
    print(f'\n⚠ Kolom tidak terdokumentasi: {kolom_tidak_terdokumentasi}')
else:
    print(f'\nSeluruh kolom sesuai dengan dokumentasi Data Dictionary.')
```

### 9.3 Validasi Tipe Data

```python
# Tipe data sesuai dokumentasi
EXPECTED_DTYPES = {
    'instant'   : 'int64',
    'dteday'    : 'object',   # Perlu konversi ke datetime pada preprocessing
    'season'    : 'int64',
    'yr'        : 'int64',
    'mnth'      : 'int64',
    'hr'        : 'int64',
    'holiday'   : 'int64',
    'weekday'   : 'int64',
    'workingday': 'int64',
    'weathersit': 'int64',
    'temp'      : 'float64',
    'atemp'     : 'float64',
    'hum'       : 'float64',
    'windspeed' : 'float64',
    'casual'    : 'int64',
    'registered': 'int64',
    'cnt'       : 'int64',
}

print('=' * 60)
print('VALIDASI TIPE DATA')
print('=' * 60)
for col, expected_type in EXPECTED_DTYPES.items():
    actual_type = str(df[col].dtype)
    status = '✅ Sesuai' if actual_type == expected_type else f'⚠ Berbeda ({actual_type})'
    print(f'  {col:<12} | Ekspektasi: {expected_type:<8} | Status: {status}')
```

### 9.4 Validasi Target Prediksi

```python
# Validasi target prediksi
EXPECTED_TARGET = 'cnt'

print('=' * 45)
print('VALIDASI TARGET PREDIKSI')
print('=' * 45)
status = '✅ Sesuai' if EXPECTED_TARGET in df.columns else '❌ Tidak Ditemukan'
print(f'Target Prediksi — Dokumentasi : {EXPECTED_TARGET} | Status: {status}')
```

### 9.5 Validasi Nilai Unik Kolom Kategorikal

```python
# Validasi nilai unik kolom kategorikal sesuai dokumentasi
EXPECTED_CATEGORIES = {
    'season'    : [1, 2, 3, 4],
    'yr'        : [0, 1],
    'mnth'      : list(range(1, 13)),
    'hr'        : list(range(0, 24)),
    'holiday'   : [0, 1],
    'weekday'   : [0, 1, 2, 3, 4, 5, 6],
    'workingday': [0, 1],
    'weathersit': [1, 2, 3, 4],
}

print('=' * 60)
print('VALIDASI NILAI UNIK KOLOM KATEGORIKAL')
print('=' * 60)
for col, expected_vals in EXPECTED_CATEGORIES.items():
    actual_vals = sorted(df[col].unique().tolist())
    sesuai = all(v in expected_vals for v in actual_vals)
    status = '✅ Sesuai' if sesuai else '⚠ Ada nilai di luar dokumentasi'
    print(f'  {col:<12} | Aktual: {actual_vals} | Status: {status}')
```

**Interpretasi:**

Hasil validasi menunjukkan:

- ✅ Jumlah baris dan kolom sesuai dokumentasi.
- ✅ Seluruh nama kolom terdokumentasi dengan benar.
- ✅ Tipe data sudah sesuai. Catatan: `dteday` bertipe `object` (string) dan akan dikonversi ke `datetime` pada tahap Data Cleaning.
- ✅ Target prediksi `cnt` terkonfirmasi.
- ✅ Nilai unik pada kolom kategorikal sesuai dengan mapping pada Data Dictionary. Jika `weathersit` hanya memiliki nilai 1–3, artinya kondisi cuaca ekstrem (nilai 4) sangat jarang atau tidak terjadi selama periode data.

**Catatan untuk Data Cleaning:**
- Kolom `dteday` perlu dikonversi dari string ke tipe `datetime`.
- Kolom `season`, `yr`, `hr`, `holiday`, `weekday`, `workingday`, dan `weathersit` secara semantik merupakan kategorikal meskipun bertipe integer.

### 10.2 Temuan Struktural

- Dataset merupakan **Multivariate Time Series** dengan granularitas per jam.
- Rentang waktu: **1 Januari 2011 – 31 Desember 2012** (2 tahun penuh).
- Kolom `dteday` bertipe `object` dan **harus dikonversi** ke `datetime` pada tahap Data Cleaning.

---

### 10.3 Temuan pada Target Prediksi

- Target prediksi adalah kolom `cnt` (total penyewaan per jam).
- Nilai `cnt` berkisar antara **1 hingga 977** dengan rata-rata sekitar **189**.
- Standar deviasi yang tinggi (~181) mengindikasikan **variasi yang besar** — analisis lebih lanjut pada EDA diperlukan untuk memahami pola distribusi.

---

### 10.4 Catatan Data Leakage

- ⚠️ Kolom `casual` dan `registered` **tidak boleh digunakan** sebagai feature input karena merupakan penyusun langsung target `cnt`.
- ⚠️ Kolom `instant` tidak memiliki nilai prediktif dan akan dihapus sebelum modeling.

---

### 10.5 Kesiapan Dataset

Dataset **siap untuk Exploratory Data Analysis (EDA)** tanpa memerlukan perbaikan kualitas data yang mendesak. Persiapan minimal yang diperlukan sebelum modeling adalah:

1. Konversi tipe data `dteday` → `datetime`.
2. Encoding kolom kategorikal jika diperlukan (sesuai metode yang dipilih).
3. Penghapusan kolom `instant`, `casual`, dan `registered` sebelum modeling.
4. Analisis dan penanganan outlier (akan dilakukan pada EDA).

---

### 10.6 Potensi Feature Engineering

Berdasarkan pemeriksaan awal, beberapa potensi feature engineering yang dapat dilakukan:

- Ekstraksi dari `dteday`: minggu ke-, kuartal, awal/akhir bulan.
- Representasi siklus untuk `hr`, `mnth`, `weekday` menggunakan fungsi sinus/kosinus.
- Lag features dari `cnt` untuk menangkap pola historis.
- Interaksi antara `hr` dan `workingday` untuk membedakan jam sibuk.
